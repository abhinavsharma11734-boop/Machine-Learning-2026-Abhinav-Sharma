## Project 1:Exploratory Data Analysis- Gapminder Datasets

#### The Gapminder dataset provides population data from 1952 to 2007 for several countries around the world. compare the populations of the European countries France,United Kingdom,Italy,Germany and spain over this period using a line chart. Make appropriate modification to the chart title, axis titles, legend, figure size, colors etc to make the chart readable and visually appealing-#Line Chart

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv(r"D:\Machine Learning\Machine-Learning-2026-Abhinav-Sharma\week-01 Intro\DATASETS\gapminderDataFiveYear.csv")


fr = df[df["country"] == "France"]
uk = df[df["country"] == "United Kingdom"]
it = df[df["country"] == "Italy"]
ge = df[df["country"] == "Germany"]
sp = df[df["country"] == "Spain"]


plt.figure(figsize=(10,6))


plt.plot(fr["year"], fr["pop"], label="France")
plt.plot(uk["year"], uk["pop"], label="United Kingdom")
plt.plot(it["year"], it["pop"], label="Italy")
plt.plot(ge["year"], ge["pop"], label="Germany")
plt.plot(sp["year"], sp["pop"], label="Spain")

plt.title("Population Comparison of European Countries (19522007)")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
plt.grid(True)

plt.show()
```

## Project 2: Exploratory Data Analysis- Gapminder Dataset
#### Diamonds_url points to a csv file containing various attributed like carat, cut, color,clarity,price etc. for over 53000 diamonds. Visualize the relationship between the carat(size of diamonds) and price using a scatter plot. Instead of using the entire dataset for this visualization, just pick the diamonds with a clarify "SI2" and color "E". use the values of the "cut" column to color the dots in the scatter plot. Make appropriate modifications to the chart title, axis titles, legend,figure size,colors etc, to make the chart readable and visually appealing. #Scatterplot.

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv(r"D:\Machine Learning\Machine-Learning-2026-Abhinav-Sharma\week-01 Intro\DATASETS\diamonds.csv")   # or diamonds_url


data = df[(df["clarity"] == "SI2") & (df["color"] == "E")]


ideal = data[data["cut"] == "Ideal"]
premium = data[data["cut"] == "Premium"]
good = data[data["cut"] == "Good"]
very_good = data[data["cut"] == "Very Good"]
fair = data[data["cut"] == "Fair"]

plt.figure(figsize=(10,6))


plt.scatter(ideal["carat"], ideal["price"], label="Ideal")
plt.scatter(premium["carat"], premium["price"], label="Premium")
plt.scatter(good["carat"], good["price"], label="Good")
plt.scatter(very_good["carat"], very_good["price"], label="Very Good")
plt.scatter(fair["carat"], fair["price"], label="Fair")


plt.title("Carat vs Price of Diamonds (Clarity: SI2, Color: E)")
plt.xlabel("Carat")
plt.ylabel("Price")
plt.legend(title="Cut")
plt.grid(True)

# Show plot
plt.show()
```

## Project 3: Exploratory Data analysis- Planets datasets
#### The Planets dataset contains details about the 1000+ extrasolar planets discovered up to 2014. Visualize the distribution of the masses of the planets (expressed as a multiple of the mass of Jupiter), using a histogram and a box plot. Make appropriate modification to the chart title, axis titles,legend,figure size, font size,colors etc. to make the chart readable and visually appealing-#Histogram and Bix Plot.

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv(r"D:\Machine Learning\Machine-Learning-2026-Abhinav-Sharma\week-01 Intro\DATASETS\planets.csv")


mass = df["mass"].dropna()


plt.figure(figsize=(10,6))
plt.hist(mass, bins=25)

plt.title("Distribution of Planet Masses", fontsize=14)
plt.xlabel("Mass (Multiple of Jupiter Mass)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(True)

plt.show()



plt.figure(figsize=(8,5))
plt.boxplot(mass)

plt.title("Box Plot of Planet Mass Distribution", fontsize=14)
plt.ylabel("Mass (Multiple of Jupiter Mass)", fontsize=12)
plt.grid(True)

plt.show()
```
