```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"D:\Machine Learning\Machine-Learning-2026-Abhinav-Sharma\week-01 Intro\DATASETS\WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
cols = [
    'Age',
    'MonthlyIncome',
    'JobSatisfaction',
    'YearsAtCompany',
    'OverTime',
    'Department',
    'JobRole',
    'WorkLifeBalance',
    'EnvironmentSatisfaction',
    'TotalWorkingYears',
    'YearsSinceLastPromotion',
    'Attrition'
]

sel_col = data[cols]
sel_col
```

```python
print("Checking Missing Values:")
print(sel_col.isnull().sum())
```

```python
print("Shape:")
print(sel_col.shape)
```

```python
print("First 15 Rows:")
sel_col.head(15)
```

```python
print("Statictics Summary:")
sel_col.describe()
```

```python
#Age
sel_col['Age'].hist()
plt.xlabel("Age")
plt.ylabel("Number of Employees")
plt.title("Age Distribution")
plt.show()

```

```python
#MonthlyIncome
sel_col['MonthlyIncome'].hist()
plt.xlabel("Monthly Income")
plt.ylabel("Number of Employees")
plt.title("Income Distribution")
plt.show()
```

```python
plt.figure()
plt.scatter(sel_col['Age'], sel_col['MonthlyIncome'])
plt.xlabel("Age")
plt.ylabel("Monthly Income")
plt.title("Age vs Monthly Income")
plt.show()
```

```python
plt.figure()

sel_col.groupby('Attrition')['MonthlyIncome'].mean().plot(kind='bar')

plt.xlabel("Attrition")
plt.ylabel("Average Income")
plt.title("Average Income vs Attrition")
```
