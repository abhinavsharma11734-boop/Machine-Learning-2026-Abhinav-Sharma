# Project 1: Exploratory Data Analysis- Gapminder Datasets


### The Gapminder dataset provides population data from 1952 to 2007 for several countries around the world. compare the populations of the European countries France,United Kingdom,Italy,Germany and spain over this period using a line chart. Make appropriate modification to the chart title, axis titles, legend, figure size, colors etc to make the chart readable and visually appealing-#Line Chart

```python

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv(r"C:\Users\abhin\OneDrive\Documents\GitHub\Machine-Learning-2026-Abhinav-Sharma\Week-01 Intro\DATASETS\gapminderDataFiveYear.csv")


countries = ["France", "United Kingdom", "Italy", "Germany", "Spain"]
eu_df = df[df["country"].isin(countries)]

plt.figure(figsize=(12, 7))


for country in countries:
    data = eu_df[eu_df["country"] == country]
    plt.plot(data["year"], data["pop"], marker='o', linewidth=2, label=country)


plt.title("Population Comparison of Major European Countries (19522007)", fontsize=15)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.legend(title="Country")
plt.grid(True, linestyle='--', alpha=0.6)

# Show plot
plt.show()
```

# Project 2: Exploratory Data Analysis- Gapminder Dataset

### diamonds_url points to a csv file containing various attributed like carat, cut, color,clarity,price etc. for over 53000 diamonds. Visualize the relationship between the carat(size of diamonds) and price using a scatter plot. Instead of using the entire dataset for this visualization, just pick the diamonds with a clarify "SI2" and color "E". use the values of the "cut" column to color the dots in the scatter plot. Make appropriate modifications to the chart title, axis titles, legend,figure size,colors etc, to make the chart readable and visually appealing. #Scatterplot.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\abhin\OneDrive\Documents\GitHub\Machine-Learning-2026-Abhinav-Sharma\Week-01 Intro\DATASETS\diamonds.csv")


filtered_df = df[(df["clarity"] == "SI2") & (df["color"] == "E")]


plt.figure(figsize=(10, 6))

for cut in filtered_df["cut"].unique():
    data = filtered_df[filtered_df["cut"] == cut]
    plt.scatter(data["carat"], data["price"], label=cut, alpha=0.7)

plt.title("Relationship Between Carat and Price of Diamonds\n(Clarity: SI2, Color: E)", fontsize=14)
plt.xlabel("Carat (Size of Diamond)", fontsize=12)
plt.ylabel("Price (USD)", fontsize=12)
plt.legend(title="Cut Type")
plt.grid(True, linestyle="--", alpha=0.5)

# Show plot
plt.show()
```

# project 3: Exploratory Data analysis- Planets datasets


### The Planets dataset contains details about the 1000+ extrasolar planets discovered up to 2014. Visualize the distribution of the masses of the planets (expressed as a multiple of the mass of Jupiter), using a histogram and a box plot. Make appropriate modification to the chart title, axis titles,legend,figure size, font size,colors etc. to make the chart readable and visually appealing-#Histogram and Bix Plot.

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv(r"C:\Users\abhin\OneDrive\Documents\GitHub\Machine-Learning-2026-Abhinav-Sharma\Week-01 Intro\DATASETS\planets.csv")


mass = df["mass"].dropna()


plt.figure(figsize=(10, 6))
plt.hist(mass, bins=30)
plt.title("Distribution of Mass of Extrasolar Planets", fontsize=14)
plt.xlabel("Planet Mass (Multiple of Jupiter Mass)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

plt.figure(figsize=(8, 5))
plt.boxplot(mass, vert=True)
plt.title("Box Plot of Planet Mass Distribution", fontsize=14)
plt.ylabel("Planet Mass (Multiple of Jupiter Mass)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
