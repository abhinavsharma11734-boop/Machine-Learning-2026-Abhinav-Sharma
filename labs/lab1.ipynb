### The Gapminder dataset provides population data from 1952 to 2007 for several countries around the world. compare the populations of the European countries France,United Kingdom,Italy,Germany and spain over this period using a line chart. Make appropriate modification to the chart title, axis titles, legend, figure size, colors etc to make the chart readable and visually appealing-#Line Chart

```python

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv(r"C:\Users\abhin\OneDrive\Documents\GitHub\Machine-Learning-2026-Abhinav-Sharma\Week-01 Intro\DATASETS\gapminderDataFiveYear.csv")


countries = ["France", "United Kingdom", "Italy", "Germany", "Spain"]
eu_df = df[df["country"].isin(countries)]

plt.figure(figsize=(12, 7))


for country in countries:
    data = eu_df[eu_df["country"] == country]
    plt.plot(data["year"], data["pop"], marker='o', linewidth=2, label=country)


plt.title("Population Comparison of Major European Countries (19522007)", fontsize=15)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.legend(title="Country")
plt.grid(True, linestyle='--', alpha=0.6)

# Show plot
plt.show()
```
